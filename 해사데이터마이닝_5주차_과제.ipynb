```python
import pandas as pd
# train = pd.read_csv("diabetes_train.csv")
# test = pd.read_csv("diabetes_test.csv")
train = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert/main/part2/ch6/diabetes_train.csv")
test = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert/main/part2/ch6/diabetes_test.csv")

# 3. 탐색적 데이터 분석(EDA)
# (1)데이터 크기 확인
print("===== 데이터 크기 =====")
print("Train Shape:", train.shape)
print("Test Shape:", test.shape)
print("\n") # 줄 바꿈

# (2)컬럼별 자료형 확인
print("===== 데이터 정보(자료형) =====")
print(train.info())
print("\n")

# (3)결측치 확인
print("===== train 결측치 수 =====")
print(train.isnull().sum().sum())
# 데이터프레임 전체의 결측치(NaN/Null) 개수를 하나의 숫자로 얻어내기 위함
print("\n")

print("===== test 결측치 수 =====")
print(test.isnull().sum().sum())
print("\n")

# (4)target 확인 -> 분류 group 개수 확인
print("===== target 빈도 =====")
print(train['Outcome'].value_counts())

target = train.pop('Outcome')

from sklearn.model_selection import train_test_split
X_tr, X_val, y_tr, y_val = train_test_split(train, target, test_size=0.2, random_state=0)

print("\n ===== 분할된 데이터 크기 =====")
print(X_tr.shape, X_val.shape, y_tr.shape, y_val.shape)

from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(random_state=0)
rf.fit(X_tr, y_tr)
pred = rf.predict_proba(X_val)

pred = rf.predict_proba(test)
submit = pd.DataFrame({'pred': pred[:,1]})
submit.to_csv("result.csv", index=False)

# 제출파일 확인
print("\n ===== 제출파일 (샘플 5개) =====")
print(pd.read_csv("result.csv").head())
```

# Type II에서 EDA가 필요한 핵심 이유
1. 데이터 상태 파악 및 오류 발견 모델 학습 전 데이터의 전반적인 상태를 이해해야 합니다. 데이터 내의 오류, 이상치, 변수 간의 관계 및 패턴을 사전에 발견하여 모델의 품질을 높일 수 있습니다.
2. 전처리 방식의 합리적 근거 제시 이후 진행될 전처리, 모델링, 평가의 방향을 ‘근거 있게’ 결정하기 위한 의사결정 단계입니다.
예를 들어, 결측치 비율을 확인하여 대체 여부를 판단하거나, 범주형 변수의 분포를 보고 적절한 인코딩 방식을 결정하는 코딩의 근거가 됩니다.
3. 모델 선택의 효율성 증대 및 시행착오 감소 데이터의 특성을 모른 채 부적절한 모델을 선택하는 비효율을 줄여줍니다. 변수 간의 선형성이나 스케일 차이를 미리 파악하여, 데이터에 적합하지 않은 모델링 시도를 방지합니다.
4. 올바른 검증 체계 및 평가 지표 설정 데이터 구조를 이해함으로써 **데이터 누수(Data Leakage)**를 피하고 검증 데이터 분할을 올바르게 수행할 수 있습니다.
특히 타깃 데이터의 불균형이 심할 경우, 단순 정확도가 아닌 F1-score나 ROC-AUC 같은 적절한 평가 지표를 우선적으로 고려하게 해줍니다.

# 요약: EDA는 단순히 데이터를 훑어보는 단계가 아니라, 전처리부터 모델링까지의 모든 과정을 논리적으로 설계하고 최적화하기 위한 필수적인 사전 설계도 역할을 합니다.
